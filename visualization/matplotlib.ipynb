```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
import numpy as np
```

```python
n = 100000
data = np.random.randn(n, 2)
```

```python
fig, ax = plt.subplots(nrows=2, figsize=(21.5, 3*2))

for i in range(2):
    ax[i].plot(np.arange(n), data[:,0], label='data')
    ax[i].legend()
```

```python
fig, ax = plt.subplots(figsize=(21.5, 3))
ax.plot(np.arange(n), data[:,0])
ax.set_xlim(100, 200)
ax.set_ylim(-10, 10)
```

```python
fig, ax = plt.subplots(figsize=(21.5, 3))
ax.plot(np.arange(n), data[:,0])
ax.get_xlim()
```

```python
fig, ax = plt.subplots(figsize=(21.5, 3))
ax.plot(np.arange(n), data[:,0])
ax.margins(-0.45, 1)
ax.get_xlim(), ax.dataLim, ax.viewLim
```
